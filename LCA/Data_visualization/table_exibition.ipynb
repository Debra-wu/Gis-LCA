```python
import matplotlib.pyplot as plt

from LCA.Data_visualization.data_processing import (
    AWE_total_emission, PEM_total_emission, SOE_total_emission,
    solar_capex_data, solar_opex_data, wind_capex_data, wind_opex_data
)

# 通用配置
bar_width = 0.35

# ========= 图一：工艺排放对比 =========
electrolyzers = ["AWE", "PEM", "SOE"]
op_emission = [AWE_total_emission, PEM_total_emission, SOE_total_emission]
construct_emission = [6.511178959100128 / 30] * 3
x = range(len(electrolyzers))

plt.figure(figsize=(8, 5))
plt.bar([i - bar_width/2 for i in x], op_emission, width=bar_width, label='Operational', color='skyblue')
plt.bar([i + bar_width/2 for i in x], construct_emission, width=bar_width, label='Construction', color='lightgreen')
plt.xticks(x, electrolyzers)
plt.ylabel("Emission (kg PM2.5 eq)")
plt.title("Emission Comparison for Hydrogen Production Processes")
plt.legend()
plt.tight_layout()
plt.show()

# ========= 图二：CAPEX 对比（solar vs wind）=========
energy_sources = ["Wind", "Solar"]
capex = [wind_capex_data, solar_capex_data]
x2 = range(len(energy_sources))

plt.figure(figsize=(6, 4))
plt.bar(x2, capex, color=['orange', 'orangered'])
plt.xticks(x2, energy_sources)
plt.ylabel("CAPEX (£/kg NH3)")
plt.title("CAPEX under Different Energy Sources")
plt.tight_layout()
plt.show()

# ========= 图三：OPEX 对比（solar vs wind）=========
opex = [wind_opex_data, solar_opex_data]

plt.figure(figsize=(6, 4))
plt.bar(x2, opex, color=['gold', 'darkorange'])
plt.xticks(x2, energy_sources)
plt.ylabel("OPEX (£/kg NH3)")
plt.title("OPEX under Different Energy Sources")
plt.tight_layout()
plt.show()
```
